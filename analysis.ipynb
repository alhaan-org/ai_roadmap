```python
import pandas as pd
df = pd.read_csv("../data/survey_results_public.csv") # In my case you can download CSV from Kaggle, Stack Overflow Survey 2025

# df.head()

south_asia_df = df[df["Country"].isin(["India", "Pakistan", "Bangladesh"])]

columns_to_keep = ["MainBranch", "Age", "EdLevel", "Employment", "WorkExp", "YearsCode", "DevType", "OrgSize", "Industry", "Country", "CompTotal", "LanguageHaveWorkedWith", "DatabaseHaveWorkedWith"]
south_asia_df = south_asia_df[columns_to_keep]

pak_devs = south_asia_df[south_asia_df["Country"] == "Pakistan"].copy()
unique_employments = pak_devs["Employment"].unique()


median_work_exp = pak_devs["WorkExp"].median()
median_comp_total = pak_devs["CompTotal"].median()
median_years_code = pak_devs["YearsCode"].median()
pak_devs["WorkExp"] = pak_devs["WorkExp"].fillna(median_work_exp)
pak_devs["YearsCode"] = pak_devs["YearsCode"].fillna(median_years_code)
pak_devs["CompTotal"] = pak_devs["CompTotal"].fillna(median_comp_total)


# print(median_work_exp)
pak_devs.to_csv("../data/pakistan_developers_according_to_stack_overflow_summary_2025.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns




freelancers = pak_devs[pak_devs["Employment"].str.contains("freelancer", na=False)]
employed = pak_devs[pak_devs["Employment"].str.contains("Employed", na=False)]
not_employed = pak_devs[pak_devs["Employment"].str.contains("Not employed|Student|I prefer not to say", na=False)]

plt.bar(employed["WorkExp"], employed["CompTotal"], color="yellow")
plt.title("Salary according to Work Experience")
plt.xlabel("Salary Range")
plt.ylabel("Years")

plt.show()

```

```python
import numpy as np
total = len(freelancers) + len(not_employed) + len(employed)

print(total)

employment_stats = np.array([len(freelancers)/total * 100, len(not_employed)/total * 100, len(employed)/total * 100])
labels = ["freelancers", "not employed", "employed"]
explode = [0.1, 0, 0]
colors = ["yellow", "red", "green"]
plt.pie(employment_stats, labels=labels, explode=explode, colors=colors)
plt.legend()
plt.title("Total Pie Chart of representing percentage of Pakistani Tech Market")
plt.show()
```
